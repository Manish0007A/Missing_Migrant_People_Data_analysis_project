```python
df.drop("Number of Children",axis=1,inplace=True)
```

```python
df.drop("Number of Females",axis=1,inplace=True)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.barh(df["Region of Incident"],df["Number of Survivors"],label=df["Number of Survivors"])
plt.yticks(c="b")
plt.show()
# It shows which region have more survivor rate.
```

```python
plt.barh(df["Region of Incident"],df["Number Dead"])
plt.yticks(c="b")
plt.show()
# it show which region is more death rate of migrant people.
```
